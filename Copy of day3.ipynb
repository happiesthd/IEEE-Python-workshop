**UNPACK TUPLE**

```python
t=('red','green','blue')

(red,green,blue)=t

print(red)
print(green)
print(blue)
```

```python
t=('red','green','blue','purple','yellow','black')

(red,green,*blue)=t

print(red)
print(green)
print(*blue)
```
