**Use of *args for variable non-keyword arguments**

```python
#Addition of variable integers
def addition(*num):
  print(type(num))
  sum=0
  for i in num:
    sum+=i
  print("The sum is :",sum)
addition(1,2,3,4,5,6,7)
print("------")
addition(7,10,13,11,17,100,1920)
```

```python
#Product of variable integers
def product(*integers):
  prd=1
  for i in integers:
    prd*=i
  print("The product is : ",prd)
product(9,58,67,223.3,14,-12.1,83,420)
```

```python
def greeting1(time,*people):
  print("Good",time," ",end="")
  for i in people:
    print(i,", ",end="")
lst= ["A","B"]
greeting1("Afternoon",*lst)
print("\n--------")
greeting1("Afternoon",lst)
print('\n--------')
greeting1("Morning",*["ABC","PQR","XYZ"],"\n")
greeting1("Morning","ABC","PQR","XYZ")
def greeting2(*people,time):
  print("Good",time," ",end="")
  for i in people:
    print(i,", ",end="")
greeting2(*["ABC","PQR","XYZ"],"Afternoon")
```

```python
def marks_add(marks):
  marks.append(55)
  print("Marks inside the function after updation : ",marks)
marks=[10,30,22,28]
marks_add(marks)
print("Marks outside the function after updation : ",marks)
```

```python
def marks_update(marks):
  grace_marks=20
  for i in marks:
    print("Original-> ",i)
    i+=grace_marks
    print("Updated->",i)
  print("Marks inside the function after updation : ",marks)
marks=[10,30,22,28]
marks_update(marks)
print("Marks outside the function after updation : ",marks)
```

```python
def marks_update(marks):
  grace_marks=20
  for i in range(len(marks)):
    print("Original->",marks[i])
    marks[i]+=grace_marks
    print("Uodated->",marks[i])
  print("Marks inside the function after updation : ",marks)
marks=[10,30,22,28]
marks_update(marks)
print("Marks outside the function after updation : ",marks)
```

**Lambda Function**

```python
def add1(x,y,z):
  return x+y+z
output1=add1(10,10,100)
print("Traditional type add1 function's output : ",output1)
print("-"*50)

add2 = lambda x,y,z : x+y+z
output2 = add2(10,10,100)
print("Lambda add2 function's output : ",output2)
print("-"*50)

add3 = lambda x,y,z : print(x+y+z)
print("Lambda add3 function's output : ",end=" ")
add3(10,10,100)
```

**Map Function**

```python
lower_to_upper_case = lambda letter : letter.upper()
lst=['a','b','C','DeF','GHI']
print("Before conversion to upper case : ",lst)
lst2=map(lower_to_upper_case,lst)
# print("After conversion to upper case : ",lst2)
print("After conversion to upper case : ",list(lst2))
```

```python
add_the = lambda string : ("The " + string)
lst=["Cat","Dog","Tiger","Lion"]
print("Before addition of the : ",lst)
lst2=map(add_the,lst)
print("After addition of the : ",list(lst2))
```

**Reduce Function**

```python
import functools
join_str = lambda x,y : x+" "+y
lst=["Cat","Dog","Tiger","Lion"]
str2=functools.reduce(join_str,lst)
print(str2)
```

**Filter Function**

```python
def even_filter(x):
  if(x%2==0):
    return True
  else:
    return False
lst=[12,33,11,98763,898,23,9]
even=filter(even_filter,lst)
print(list(even))
```

```python
def eligible(inp):
  if(inp[1]>18):
    return True
  else:
    return False
info={"Rohan":16,"Rakshita":19,"Shreya":22}
eligible_stud=dict(filter(eligible,info.items()))
print(eligible_stud)
```

**Recursion**

```python
#Factorial of a number using recursion
```

```python
def fibo(n):
   if n <= 1:
       return n
   else:
       return(fibo(n-1) + fibo(n-2))

nterms = 10

if nterms <= 0:
   print("Please enter a positive integer !!")
else:
   print("Fibonacci sequence:")
   for i in range(nterms):
       print(fibo(i))
```
