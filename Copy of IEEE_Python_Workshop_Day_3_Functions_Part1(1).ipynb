# **Variable Scopes**

The scope of a variable determines the areas where it is accessible.

**Q. What's a namespace ?**

-> A namespace is a system that has a unique name for each and every object in Python. An object might be a variable or a function. 

```python
# var1 is in the global namespace
var1 = 5
def function1():
    print(var1)
    # var2 is in the local namespace
    var2 = 6
    
    def function2():
        print(var2)
        # var3 is in the nested local namespace
        var3 = 7
        print(var3)
    function2()

function1()
```

```python
def myfunc():
  x = 300
  def myfunc2():
    print(x)
  myfunc2()

myfunc()
```

```python
global_var = 10

def my_function():
    # defined a local variable
    local_var = 20
 
    global global_var
    global_var = 30

# print global variable value
print(global_var)

# call the function and modify the global variable
my_function()

# print the modified value of the global variable
print(global_var)
```

## **Types of functions based on parameters declared and return value**

1) With parameters and return value

2) With parameters and without return value

3) Without parameters and with return value

4) Without parameters and return value



###  1) With parameters and return value

```python
#1) With parameters and return value
def add(a, b):
    addition = a + b
    return addition
# We are calling it Outside the Definition
print("After Calling :", add(25, 45))
```

###  2) With parameters and without return value

```python
#2) With parameters and without return value
def multi(a, b):
    mult = a * b
    print("After Calling:", mult)
# Same here, outside the Definition
multi(10, 20)   
```

### 3) Without parameters and with return value

```python
#3) Without parameters and with return value
def Multiplication():
    a = 10
    b = 25
    Multi = a * b
    return Multi
#Here too, we are calling it outside the Definition    
print("After Calling the Multiplication : ", Multiplication())
```

### 4) Without parameters and return value

```python
#4) Without parameters and return value
def Addition():
    a = 20
    b = 30
    sum = a + b
    print("After Calling :", sum)
#Same here too.
Addition()
```

## **Positional, Default, and Keyword Arguments**
1) **Positional** argument values are assigned according to the **sequence**.

2) **Keyword** argument values get assigned according to their keyword during argument passing.

3) **Default** arguments assign default values at the time of function **definition**.

```python
def my_sum (a,b,):
  return a+b
my_sum(10,20)
```

2) **Keyword** arguments are values that are assigned according to their keyword during argument passing.

```python
def my_sum (a,b):
  return a+b
my_sum(a=30, b=25)
```

3) **Default** arguments assign default values at the time of function definiton.

```python
def my_sum (a,b=40):
  return a+b
my_sum(10)
```

Let's test this out! 

```python
def message(name, surname):
    print("Hello", name, surname)

message(name="John", surname="Wilson")
message(surname="Jackson", name="Michael")
```

In keyword arguments order of argument does not matter, but the number of arguments must match. Otherwise, we will get an error.

###**NOTE:** 
While using keyword and positional argument simultaneously, we need to pass 1st arguments as positional arguments and then keyword arguments. Otherwise, we will get SyntaxError.

```python
def message(first_nm, last_nm):
    print("Hello..!", first_nm, last_nm)

# Correct Use.
message("Steve", "Jobs")
message("Steve", last_nm="Jobs")
```
